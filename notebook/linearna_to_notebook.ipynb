```
# =============================================================================
# ĆELIJA 1: Uvoz biblioteka i učitavanje podataka
# =============================================================================
import pandas as pd
import numpy as np

# Učitavamo Boston dataset (Pazi na putanju, stavi '../data/boston.csv' ako ti je data folder nazad)
data = pd.read_csv('../data/boston.csv')
print("Kako izgledaju naši sirovi podaci:")
data.head()
```

```
# =============================================================================
# ĆELIJA 2: Izdvajanje X (atributa) i y (ciljne cene)
# =============================================================================
target_col = data.columns[-1]
print(f"Kolona koju predviđamo: {target_col}\n")
```

```
y = data[[target_col]].to_numpy()
y
```

```

X_raw = data.drop(columns=[target_col])

print(f"Dimenzije ciljnih cena y: {y.shape}")
print(f"Dimenzije ulaznih atributa X: {X_raw.shape}")
X_raw
```

```
# =============================================================================
# ĆELIJA 3: KORAK 1 - Normalizacija (Z-Score)
# =============================================================================
# Zašto ovo radimo? Da bismo sve kolone sveli na sličan raspon. Gradijentni spust
# inače ne radi dobro ako su neki brojevi ogromni (npr. porez), a neki mali (broj soba).

X_mean = X_raw.mean()
# 'replace' rešava granični slučaj: sprečava deljenje sa nulom ako su svi podaci isti!
X_std = X_raw.std().replace(0, 1e-9)

# x mean i std su panda series - 1d niz sa indeksom(nazivom)

X_norm = (X_raw - X_mean) / X_std

# sve ih dovodi u raspon npr - 2 do 2, za koliko koraka tj st devijacija je ova vrednost udaljena od proseka

print("Podaci NAKON normalizacije (sada su oko nule):")
X_norm.head()
```

```
# =============================================================================
# ĆELIJA 4: KORAK 2 - Dodavanje Bias-a (Slobodnog člana)
# =============================================================================
# Dodajemo kolonu sa jedinicama na KRAJ matrice, kako bi se slobodni član w0 množio sa 1.
X_norm_np = X_norm.to_numpy()
X_norm_np
# ovde smo samo pretvorili u np niz - 2d , svaki unutrasnji niz su samo redovi kuca sa vrednostima iz kolona
# jer navodno nam smetaju imena kolona za matricno mnozenje i guess
```

```

m = X_norm_np.shape[0] # vraca broj redova - 506

X = np.hstack([X_norm_np, np.ones((m, 1))]) # mergeuje 2 matrice np(ones) - koja je 506x1 i x_norm_Np

print(f"Dimenzije matrice X pre biasa: {X_norm_np.shape}")
print(f"Dimenzije matrice X POSLE biasa: {X.shape} (Dodata 1 kolona)")
print(f"Prvi red podataka sada izgleda ovako (uoči jedinicu na kraju!):\n{X[0]}")
```

```
# =============================================================================
# ĆELIJA 5: MATEMATIKA - Jedan korak gradijentnog spusta (Pešaka)
# =============================================================================
n = X.shape[1] # broj parametara (14)

# Inicijalizujemo male nasumične parametre
w = np.random.randn(1, n) * 0.01

# randn je za normalnu Gausovu - brojevi blizu nule u dimenziji 1x14
# i jos ih dodatno smanjujemo sa 0.01 jer mogu ici i do 2 i 3

print(f"Početne težine (w) pre učenja:\n{w}\n")

# A) PREDVIĐANJE: Množimo podatke X sa transponovanim težinama w
pred = X @ w.T
pred[1]

# transponovali smo jer je matrica bila 14x1 a zelimo da bude 1x14 zbog 506x14

# za prvi slucaj tj red svaku kolonu mnozi sa odgovarajucom tezinom tj redom iz vektora w
# i na kraju za ceo taj red sabira sve te pomnoske i dobije jedan broj za tu pred kolonu

# pred je sada 1 kolona sa 506 redova naseg predvidjanja
```

```

# B) GREŠKA: Razlika između našeg predviđanja i pravih cena
err = pred - y
err
# greska je opet 1 kolona sa 506 redova
```

```

# C) GRADIJENT: Pravac u kom treba da pomerimo težine da bismo smanjili grešku
# delimo sa 506 tj brojem uzoraka jer je to prosecna greska

# znaci err smo vec rekli to je greska za svaki red tj svaku kucu
# onda tu gresku za taj jedan red mnozimo sa kolonom iz date i sabiramo znaci e1xCRIM1 + e2xCRIM2 ...
grad = (err.T @ X) / m
grad
```

```

# D) REGULARIZACIJA: Kaznjavamo velike težine, ali NE DIRAMO w0 (poslednji indeks -1)

# postavljamo regularizacioni koeficijent lambda
# znaci da ce u onom azuriranju toliko biti smanjene tezine
lam = 0.1
reg = lam * w.copy()
reg[0, -1] = 0.0
# 0 je prvi red (nas jedini iz 1x14) a -1 je za bias da mu kazna bude 0

grad = grad + reg

# E) AŽURIRANJE: Pomeramo w u pravcu "nizbrdice" (suprotno od gradijenta)

# alfa je learning rate
alpha = 0.01
w = w - alpha * grad

print(f"Težine (w) NAKON jedne iteracije učenja (malo su se promenile!):\n{w}")
```

```
# =============================================================================
# ĆELIJA 6: SPAKUJMO SVE U KLASNE FUNKCIJE
# =============================================================================
def normalize(X):
    mean = X.mean()
    std = X.std().replace(0, 1e-9)
    X_norm = (X - mean) / std
    return X_norm, mean, std

def add_bias(X):
    m = X.shape[0]
    return np.hstack([X, np.ones((m, 1))])

def learn(data, target_col, lam=0.0, alpha=0.01, max_iter=10000, tol=0.01, method='batch', auto_lr=False, verbose=True):
    y = data[[target_col]].to_numpy()
    X_raw = data.drop(columns=[target_col])

    X_norm, X_mean, X_std = normalize(X_raw)
    X = add_bias(X_norm.to_numpy())
    m, n = X.shape

    w = np.random.randn(1, n) * 0.01

    prev_w = None
    prev_grad = None

    for it in range(max_iter):
        pred = X @ w.T
        err = pred - y

        if method == 'sgd':
            # SGD: nasumično biramo samo 1 kuću za proračun!
            idx = np.random.randint(m)
            grad = err[idx] * X[idx].reshape(1, -1)
        else:
            # BATCH: koristimo celu bazu
            grad = (err.T @ X) / m

        # L2 regularizacija — ne penalizujemo slobodni član
        reg = lam * w.copy()
        reg[0, -1] = 0.0
        grad = grad + reg

        # Barzilai-Borwein automatski learning rate (BONUS)
        if auto_lr and method == 'batch' and prev_w is not None:
            delta_w = w - prev_w
            delta_grad = grad - prev_grad
            denom = float(delta_grad @ delta_grad.T)
            if denom > 1e-12:
                alpha = abs(float(delta_w @ delta_grad.T) / denom)

        prev_w = w.copy()
        prev_grad = grad.copy()
        w = w - alpha * grad

        grad_norm = float(np.abs(grad).sum())
        loss = float(np.mean(err ** 2))

        # Ispis svaki 500-ti korak
        if verbose and it % 500 == 0:
            print(f"iter={it:5d} | grad_norm={grad_norm:.6f} | MSE={loss:.4f} | alpha={alpha:.6f}")

        # Uslov za prevremeni izlaz (ako smo stigli do dna brda)
        if grad_norm < tol:
            if verbose:
                print(f"\nKonvergencija u iteraciji {it}")
            break

    return {'w': w, 'X_mean': X_mean, 'X_std': X_std}

def predict(model, data_new):
    # Nove podatke moramo normalizovati starim X_mean i X_std!
    X_norm = (data_new - model['X_mean']) / model['X_std']
    X = add_bias(X_norm.to_numpy())
    return (X @ model['w'].T).flatten()
```

```
# =============================================================================
# ĆELIJA 9: TEST 3 - TRENIRANJE (Bonus Automatska Alfa)
# =============================================================================
print(">> POKREĆEM BATCH SA AUTOMATSKIM LEARNING RATE-OM <<")
# Pazi koliko će iteracija ovome trebati da konvergira! Skoro trenutno!
model_auto = learn(data, target, lam=0.1, method='batch', auto_lr=True)
```

```
# =============================================================================
# ĆELIJA 10: PREDIKCIJA I EVALUACIJA MODELA
# =============================================================================
# Koristićemo najbolji model (auto) za predikciju na svim podacima
X_test = data.drop(columns=[target])
y_true = data[target].to_numpy()

# Pozivamo funkciju predict
preds = predict(model_auto, X_test)

# Vizuelni prikaz: Prvih 5 predikcija vs Stvarne cene
print("Naša predikcija\t| Stvarna cena")
print("-" * 35)
for i in range(5):
    print(f"{preds[i]:.2f}\t\t| {y_true[i]:.2f}")

# Računamo ukupnu RMSE (Root Mean Square Error) grešku
rmse = np.sqrt(np.mean((preds - y_true) ** 2))
print(f"\nKonačna prosečna greška (RMSE) modela je: {rmse:.4f}")
```
